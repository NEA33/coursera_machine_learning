```python
import pandas
import numpy as np
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
data = pandas.read_csv('wine.data', names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
                                             'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
                                             'Proanthocyanins', 'Color intensity', 'Hue',
                                             'OD280/OD315 of diluted wines', 'Proline'])
```

```python
wine = data[['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash','Magnesium', 
             'Total phenols', 'Flavanoids', 'Nonflavanoid phenols','Proanthocyanins', 'Color intensity', 
             'Hue', 'OD280/OD315 of diluted wines', 'Proline']]
y = data[['Class']]
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 42)
scoreK = np.zeros(50)
for train_index, test_index in kf.split(wine):
    X_train, X_test = wine.loc[train_index], wine.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    for k in range(1, 51):
        neigh = KNeighborsClassifier(n_neighbors = k)
        neigh.fit(X_train, y_train.values.ravel())
        scoreK[k-1] += neigh.score(X_test, y_test)
scoreK /= 5
```

```python
plt.plot(scoreK)
```

```python
# k
k = scoreK.argmax() + 1
with open('week2_1', 'w') as f:
    f.write(str(k))

with open('week2_2', 'w') as f:
    f.write(str(round(scoreK[k-1], 2)))
    
print(k)
print(round(scoreK[k-1], 2))
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 42)
scoreK_scaled = np.zeros(50)
wine_scaled = preprocessing.scale(wine)
for train_index, test_index in kf.split(wine_scaled):
    X_train, X_test = wine_scaled[train_index], wine_scaled[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    for k in range(1, 51):
        neigh = KNeighborsClassifier(n_neighbors = k)
        neigh.fit(X_train, y_train.values.ravel())
        scoreK_scaled[k-1] += neigh.score(X_test, y_test)
scoreK_scaled /= 5
```

```python
plt.plot(scoreK_scaled)
```

```python
# k_scaled
k = scoreK_scaled.argmax() + 1
with open('week2_3', 'w') as f:
    f.write(str(k))

with open('week2_4', 'w') as f:
    f.write(str(round(scoreK_scaled[k-1], 2)))
    
print(k)
print(round(scoreK_scaled[k-1], 2))
```

```python
#5
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X = preprocessing.scale(boston.data)
y = boston.target
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 42)
p = np.linspace(1, 10, 200)
scoreK_scaled = np.zeros(200)
for k in range(200):
    neigh = KNeighborsRegressor(n_neighbors=5, weights = 'distance', p = p[k], metric='minkowski')
    scoreK_scaled[k] += (cross_val_score(estimator=neigh, X=X, y=y,
                                         scoring='neg_mean_squared_error',
                                         cv = kf)).mean()
```

```python
plt.plot(scoreK_scaled)
```

```python
with open('week2_5', 'w') as f:
    f.write(str(round(p[scoreK_scaled.argmax()],1)))
```

```python
#6
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
train = pandas.read_csv('perceptron-train.csv', names = ['y', 'x1', 'x2'])
test = pandas.read_csv('perceptron-test.csv', names = ['y', 'x1', 'x2'])
```

```python
y_train = train['y']
X_train = train[['x1','x2']]

y_test = test['y']
X_test = test[['x1', 'x2']]
```

```python
clf = Perceptron(random_state=241)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
score1 = accuracy_score(y_test, predictions)
score1
```

```python
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
clf = Perceptron()
clf.fit(X_train_scaler, y_train)
predictions = clf.predict(X_test_scaler)
score2 = accuracy_score(y_test, predictions)
score2
```

```python
with open('week2_6', 'w') as f:
    f.write(str(round(score2-score1,3)))
```
